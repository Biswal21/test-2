```python
import re

speaker_split_text = []

for index,row in df.iterrows():
  # Initialize variables
  speakers = {}
  current_speaker = None
  current_paragraph = ""

  # Split the transcript by lines and identify speakers
  lines = row['Prepared Remarks'].split('\n')
  for line in lines:
      speaker_match = re.match(r"^(.*?) -- (.*)$", line)
      if speaker_match and f"{speaker_match.group(1)} -- {speaker_match.group(2)}" in row['Call participants']:
          current_speaker = f"{speaker_match.group(1)} -- {speaker_match.group(2)}"
          current_paragraph = ""
          speakers.setdefault(current_speaker, [])
      else:
          if current_speaker:
              speakers[current_speaker].append(line)


  for speaker, paragraphs in speakers.items():


    # paragraphs_comb = [paragraphs[i]+" "+paragraphs[i + 1] if i + 1 < len(paragraphs) else paragraphs[i] for i in range(0, len(paragraphs), 2)]
    pattern = r'^(.*?)\s*--\s*(.*?)$'

    match = re.search(pattern, speaker)

    if match:
      name = match.group(1)
      designation = match.group(2)


    for paragraph in paragraphs:
      speaker_split_text.append({'Company':row['Company'],'Year':row['year'],'Quarter': row['quarter'],'Speaker':name,'Designation':designation,'Paragraph':paragraph})


Speaker_df = pd.DataFrame(speaker_split_text)
```

```python
Speaker_df = Speaker_df.fillna(0)
```

```python
data_types = {
    'Year': int,
    'Quarter': int
}

Speaker_df = Speaker_df.astype(data_types)
```

```python
B_INST, E_INST = "[INST]", "[/INST]"
B_SYS, E_SYS = "<<SYS>>\n", "\n<</SYS>>\n\n"
# DEFAULT_SYSTEM_PROMPT = """\
# You are a helpful, respectful and honest assistant. Always answer as helpfully as possible, while being safe. Your answers should not include any harmful, unethical, racist, sexist, toxic, dangerous, or illegal content. Please ensure that your responses are socially unbiased and positive in nature.

# If a question does not make any sense, or is not factually coherent, explain why instead of answering something not correct. If you don't know the answer to a question, please don't share false information."""

def get_prompt(instruction, new_system_prompt):
    SYSTEM_PROMPT = B_SYS + new_system_prompt + E_SYS
    prompt_template =  B_INST + SYSTEM_PROMPT + instruction + E_INST
    return prompt_template
```

```python
template = """
[INST]
<<SYS>>
You are an Earnings Calls Analyzer, you read the text given by the user in an analytical manner and give the top 5 major topics talked about in that text.
Here are a few guidelines for generating topics:

1. Topics generated should not contain positions such as CEO, CFO,etc.

2. Topics also should not contain names of people.

3. Whenever possible give a topic that is more specific than general.

4. Keep topics concise rather than verbose.

5. Keep the topics to 4 to 5 words max
<</SYS>>

Provided below is the chat history with you.(Ignore if none given)
{chat_history}

Read the paragraph given below and give the major 5 topics in the paragraph as a comma-separated list inside square brackets. Example: [t1, t2, t3].
The examples given in chat history should be looked at for output format if provided.

Human: {text}
Output:
[/INST]
"""

prompt = PromptTemplate(
    input_variables=["chat_history", "text"], template=template
)
```
